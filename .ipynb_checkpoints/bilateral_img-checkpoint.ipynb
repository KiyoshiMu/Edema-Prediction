```python
%matplotlib inline
from collections import defaultdict
import pandas as pd
from tools import load_pickle, clean_df, markdown_tabel
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind, pearsonr
from sklearn.preprocessing import StandardScaler
import numpy as np
sns.set()
```

First, we selected qualified samples, whose left & right both meet the requirement, that is the volume is larger than 2000 $mm^3$. Also, two datestamps, i.e. before and after therapy, are required.

```python
def have4time(index_list):
    container = defaultdict(list)
    result = []
    for index in index_list:
        name = index.split('Date')[0]
        container[name].append(index)
        if len(container[name]) == 4:
            result.extend(container[name])
    return result
```

```python
def split_pipe(data, selected_feature, key='original_shape_Volume'):
    data = data.sort_index()
    split_left = data.iloc[::2,:]
    split_right = data.iloc[1::2,:]
    x_left, y_left = clean_df(split_left, selected_feature, key)
    x_right, y_right = clean_df(split_right, selected_feature, key)
    return x_left, y_left, x_right, y_right
```

```python
def read_bi_pipe(p=r'data/split_img.xlsx', 
                 pkl_p='data/selected_feature.pkl', key='original_shape_Volume'):
    split_data_temp = pd.read_excel(p)
    # the following step may casue the loss of data integrity, 'have4time' can make sure data is intact
    split_data = split_data_temp.loc[split_data_temp[key]>2000]
    sel_split_temp = split_data.loc[have4time(split_data.index.tolist())]
    selected_feature = load_pickle(pkl_p)
    x_left, y_left, x_right, y_right = split_pipe(sel_split_temp, selected_feature, key=key)
    return x_left, y_left, x_right, y_right
```

Here, we got desired samples and features.

```python
x_left, y_left, x_right, y_right = read_bi_pipe()
```

```python
def y_creator(ori_y_p, sel):
    y = pd.read_excel(ori_y_p, index_col=0)
    y.index = y.index.astype(str)
    y_ori = y.loc[sel,:]
    return y_ori
```

```python
y_ori = y_creator('data/y.xlsx', x_left.index)
```

```python
def reliability_test(y, y_left, y_right):
    v_bind = y_left.iloc[:, :2] + y_right.iloc[:, :2]
    gap = abs(v_bind - y.iloc[:, :2]) / y.iloc[:, :2]
    return gap
```

```python
rt_outcome = reliability_test(y_ori, y_left, y_right)
```

Next, we categorized the samples by whether they have discrepancy of therapy effectiveness between in parts and in whole. If the edema volume in left shrinked over $25%$, while the volume in right remained the same or even rised, this sample belongs to different outcome group. Besides, in the preceding case, if the whole edame volume became smaller than $75%$ of that before the therapy and the edema volume in left is larger than that in right, this case is dominated by large part.

```python
def pie_chart(numbers, labels, name='pics/supplement-figure1.png'):
    def func(pct, allvals):
        absolute = int(pct/100.*np.sum(allvals))
        return "{:.1f}%\n({:d})".format(pct, absolute)
    
    colors = ['#ff9999','#66b3ff','#ffcc99']
    explode = (0, 0, 0)
    fig, ax = plt.subplots()
    ax.pie(numbers, explode=explode, labels=labels, colors=colors, autopct=lambda pct: func(pct, numbers),
            startangle=90)
    # Equal aspect ratio ensures that pie is drawn as a circle
    #draw circle
    centre_circle = plt.Circle((0,0),0.70,fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)
    
    ax.axis('equal')
    plt.tight_layout()
    plt.savefig(name, dpi=300, bbox_inches="tight")
```

```python
def pie_helper(over_same, large_dominate):
    amount = len(over_same)
    num_same = sum(over_same)
    num_diff = amount - num_same
    num_ds = amount - sum(large_dominate)
    num_dl = num_diff -num_ds
    print(f"""Different Outcomes and Dominated by Large Part {num_dl},
         Different Outcomes and Dominated by Small Part,{num_ds},
         Same Outcomes: {num_same}""")
    
    pie_chart([num_dl, num_ds, num_same],
         ['Different Outcomes and Dominated by Large Part',
         'Different Outcomes and Dominated by Small Part',
         'Same Outcomes'])
```

```python
def trait_extract(y, y_left, y_right):
    """compute the realationship between bilateral parts and the whole brain"""
    left_same = y_left['effectiveness'] == y['effectiveness']
    right_same = y_right['effectiveness'] == y['effectiveness']
    over_same = left_same == right_same
    # test the reliability
    lr_same = y_left['effectiveness'] == y_right['effectiveness']
    assert sum(lr_same==over_same)/len(over_same) == 1, "Please Check the data, it is not consistent"
    left_large = y_left['volume before'] > y_right['volume before']
    left_dominate = left_large == left_same
    # when the right and the left are different, whelther the larger one play a more important role
    large_dominate = over_same | left_dominate
    df = pd.DataFrame({'over_same':over_same, 'fit_left':left_same, 'fit_right':right_same, 'large_dominate':large_dominate})
    pie_helper(over_same, large_dominate)
    return df
```

```python
comparison = trait_extract(y_ori, y_left, y_right)
```

Here, we computed the distance in one patient (the distance between one patient's two parts) and the distance in different patients (the distance between one patient and the other patient) of each feature. Distances are judged by absolute difference.

```python
def ingroup_distance(g1, g2, c):
    return abs(g1 - g2).values.tolist()
```

```python
def outitem_distance(g1_sum, g2_sum, g1_len, g2_len, item):
    return  abs(g1_sum - item * g1_len) + abs(g2_sum - item * g2_len)
```

```python
def outgroup_distance(g1, g2, c):
    g1_sum, g2_sum, g1_len, g2_len = sum(g1), sum(g2), len(g1), len(g2)
    return [(outitem_distance(g1_sum, g2_sum, g1_len, g2_len, item1) + \
             outitem_distance(g1_sum, g2_sum, g1_len, g2_len, item2)) / (2 * (g1_len + g2_len - 2))
            for item1, item2 in zip(g1, g2)]
```

```python
def distance_pipe(df1, df2):
    recoder = {}
    for c in df1.columns:
        g1 = df1[c]
        g2 = df2[c]
        ingroup = ingroup_distance(g1, g2, c)
        outgroup = outgroup_distance(g1, g2, c)
        recoder[c] = (ingroup, outgroup)
    return recoder
```

```python
def distance_plot(distance_recoder, name='pics/figure4.png'):
    fig, axes = plt.subplots(2, 3,figsize=(15,8))
    axes_f = axes.flatten()
    sns.set()
    for idx, (k, v) in enumerate(distance_recoder.items()):
        ax = axes_f[idx]
        ingroup = v[0]
        outgroup = v[1]
        
        sns.distplot(ingroup, ax = ax, label='in One Patient')
        sns.distplot(outgroup, ax = ax, label='in Different Patients')
        ax.set_title(k)
        p = ttest_ind(ingroup, outgroup)[1]
        ax.text(0.76, 0.15, "p-value = {:.4f}".format(p), horizontalalignment='center',
            verticalalignment='center',transform = ax.transAxes, fontsize="medium")
#         ax.set_xlabel('Distence of Radiomics Image Features')
#         ax.set_ylabel('Density')

        ax.legend()
    plt.text(-5.5, -0.5, 'Absoulte Difference', fontsize="medium")
    plt.text(-14, 2.3, 'Density', fontsize="medium", rotation=90)
    plt.savefig(name, dpi=300, bbox_inches="tight")
```

```python
def normalized(x):
    scaler = StandardScaler()
    x_out = scaler.fit_transform(x)
    return x_out
```

```python
def ingroup_distance_whole(g_left, g_right, ratio):
    gap = (g_left - g_right) ** 2 * ratio
    out = np.sum(gap, axis=1)
    return out
```

```python
def stagger(item, group, ratio):
    container = []
    for comp in group:
        item_distance = sum((item - comp) ** 2 * ratio)
        container.append(item_distance)
    one = sum(container)
    return one
```

Next, we combined 6 features to compare the distance in one patient and that in different patients. This distance can be explained by the following case. Assuming a is in the left group, b is the right group, the distance is the mean of all distances that are between a and the others in left group, between a and all parts in right except the one belonging to the same sample, between b and the others in right group, between b and all parts in left except the one belonging to the same sample.

```python
def outgroup_distance_whole(g_left, g_right, ratio, ingroup_distance):
    left_lefts = [stagger(left, g_left, ratio) for left in g_left]
    left_rights = [stagger(left, g_right, ratio) for left in g_left]
    right_lefts = [stagger(right, g_left, ratio) for right in g_right]
    right_rights = [stagger(right, g_right, ratio) for right in g_right]
    result = (np.array(left_lefts) + np.array(left_rights) + np.array(right_lefts) + np.array(right_rights) -\
             (ingroup_distance * 2)) / (len(g_left) * 4 - 4)
    return result
```

```python
def distance_whole_pipe(x_left, x_right, pkl_p='data/adjust_ratio.pkl'):
    ratio_dict = load_pickle(pkl_p)
    ratio = [ratio_dict[c] for c in x_left.columns]
    g_left = normalized(x_left)
    g_right = normalized(x_right)
    ingroup_distance = ingroup_distance_whole(g_left, g_right, ratio)
    outgroup_distance = outgroup_distance_whole(g_left, g_right, ratio, ingroup_distance)
    plot_distance_whole(ingroup_distance, outgroup_distance)
```

```python
def plot_distance_whole(in_one, in_diff, name='pics/figure5.png', l_first="in One Patient",
                       l_second='in Different Patients', xlabel='Distance of Radiomics Image Features'):
    fig, ax = plt.subplots()
    sns.distplot(in_one, label=l_first)
    sns.distplot(in_diff, label=l_second)
    p_val = ttest_ind(in_one, in_diff)[1]
    ax.text(0.8, 0.35, "p-value = {:.4f}".format(p_val), horizontalalignment='center',
            verticalalignment='center',transform = ax.transAxes, fontsize="medium")
    plt.xlabel(xlabel)
    plt.ylabel('Density')
    plt.legend()
    plt.savefig(name, dpi=300, bbox_inches="tight")
```

Finally, we compared the correlation between left features and right features in same outcome group and different outcome group. If the correlation of one feature is high in the former group, while it's low in the latter group, this feature may indicate some peculiarities of the difference, which is worth studying.

```python
def plot_bi(df_set, name='pics/figure6.png'):
    sns.set()
    fig, axes = plt.subplots(6, 2,figsize=(8,24))
    axes_f = axes.flatten()
    archive = [{},{}]
    for j, dfs in enumerate(df_set):
        left = dfs[0]
        right = dfs[1]
        xs = left
        ys = right
        
        for i, c in enumerate(xs.columns):
            x = xs[c]
            y = ys[c]
            archive[j][c] = list(pearsonr(x, y))
            
            if j == 0:
                ax = axes_f[i*2]
            else:
                ax = axes_f[i*2 + 1]
            ax.scatter(x, y)

            ax.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)),":" )
            correlation = np.corrcoef(x, y)[0,1]
            rsq = correlation ** 2

            ax.text(0.67, 0.1, "    $R^2$ = {:.4f}".format(rsq), horizontalalignment='center',
            verticalalignment='center',transform = ax.transAxes, fontsize="x-small")

            ax.set(xlabel="LEFT" , ylabel='RIGHT')
            ax.set_title(c)
    fig.tight_layout()
    plt.text(-0.15, 14, 'Same', fontsize="large")
    plt.text(2.7, 14, 'Different ', fontsize="large")
    plt.savefig(name, dpi=300, bbox_inches="tight")
    return archive
```

```python
def corr_bi_pipe(comparison, x_left, x_right):
    same = comparison.loc[comparison['over_same'] == True].index
    different = comparison.loc[comparison['over_same'] != True].index
    df_set = ((x_left.loc[same], x_right.loc[same]), (x_left.loc[different], x_right.loc[different]))
    return plot_bi(df_set)
```

```python
peculiarity = corr_bi_pipe(comparison, x_left, x_right)
```

```python
def show_peculiarity(pec_dict):
    def show(part_dict):
        df = pd.DataFrame.from_dict(part_dict, orient='index', columns=["Pearson correlation coefficient", 'p-value'])
        return df
    same_df = show(pec_dict[0])
    markdown_tabel(same_df, '.')
    diff_df = show(pec_dict[1])
    markdown_tabel(diff_df, '.')
    print('same')
    print(same_df)
    print('different')
    print(diff_df)
```

```python
def one_feature(comparison, x_left, x_right, y_left, y_right, feature):
    different = comparison.loc[comparison['over_same'] != True].index
    g_y = y_left.loc[different].append(y_right.loc[different])
    g_x = x_left.loc[different].append(x_right.loc[different])
    g1 = g_x.loc[g_y['effectiveness'] == True, feature]
    g2 = g_x.loc[g_y['effectiveness'] == False, feature]
    plot_distance_whole(g1, g2, name='pics/figure7.png', l_first="Effective",
                       l_second='Ineffective')
```
